Renomeação, tradução e criação de uma coluna e inserida novamente no dataframe


```python
#pd.set_option("display.max_rows", None)
#print(df['TipoDev'].value_counts().head(100))
```

```python
df_tipodev = df[['TipoDev']].copy()
termos_tipodev = ['cientista de dados','fullstack','backend','frontend']

df_tipodev.loc[df_tipodev['TipoDev'].str.contains('Data',  flags=re.IGNORECASE, regex=True,na = False)] = 'cientista de dados'
df_tipodev.loc[df_tipodev['TipoDev'].str.contains('full-stack',  flags=re.IGNORECASE, regex=True, na = False)] = 'fullstack'
df_tipodev.loc[df_tipodev['TipoDev'].str.contains('back-end',  flags=re.IGNORECASE, regex=True, na = False)] = 'backend'
df_tipodev.loc[df_tipodev['TipoDev'].str.contains('front-end',  flags=re.IGNORECASE, regex=True, na = False)] = 'frontend'
df_tipodev.loc[~df_tipodev['TipoDev'].str.contains('|'.join(termos_tipodev), na = True)] = 'outro'
df_tipodev
```

```python
df.drop(['TipoDev'], axis=1, inplace=True)
```

```python
df = pd.concat([df, df_tipodev], axis=1)
```
